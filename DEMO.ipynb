# Welcome

ModerAItor is a tool to automatically moderate comments on public forums.

# Overview

Public forums are places where people freely share their opinion. However, this
comes with a cost, toxicity and false information thrive in this places which
is very harmful for society:

Moreover, moderating comments is a Herculian task, it must be automatized.
In a popular place such as Reddit, [there are billions of comments to review per
year](https://www.redditinc.com/blog/reddits-2020-year-in-review/):



Moreover, the moderators are completely voluntary, so ModerAItor can make
their life easier.

# What I used

- [LangChain](https://python.langchain.com/en/latest/index.html)
- Large Language Model (LLM): [OpenAssistant/oasst-sft-4-pythia-12b-epoch-3.5](https://huggingface.co/OpenAssistant/oasst-sft-4-pythia-12b-epoch-3.5) + HuggingFace Inference API
- [FastAPI](https://fastapi.tiangolo.com/)


# Demo

- Choose a post: https://www.reddit.com/r/OldSchoolCool/comments/14a3n3y/how_a_mugshot_of_a_criminal_looked_like_in_1925/
- Choose the rules: "No offensive comments, offensive submissions or repost spam"

```python
%%!

(
curl -X POST -H "Content-Type: application/json" -d '{
  "url": "<URL>",
  "rules": "<RULES>",
  "max_comments": 10 
}' http://54.147.130.67/api/moderate_post/ >> output.json
)
```

```python
import json

with open("output.json", "r") as f_hdl:
    result = json.load(f_hdl)

print(json.dumps(result, indent=4))
```
